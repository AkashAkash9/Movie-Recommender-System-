```python
import pandas as pd
```

```python
# Load datasets
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

Create the User-Movie Matrix

Rows = users

Columns = movies

Cells = ratings (0–5 stars)

```python
# Merge ratings with movie titles
data = pd.merge(ratings, movies, on='movieId')

# Create user-movie matrix (users as rows, movie titles as columns)
user_movie_matrix = data.pivot_table(index='userId', columns='title', values='rating')

user_movie_matrix.head()
```

Build a Movie-to-Movie Recommender

```python
#Pick a movie (e.g., Toy Story)
#Calculate its similarity with every other movie based on user ratings
#Return the top most similar movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Fill NaN with 0 (assume unrated movies = 0 rating)
movie_user_matrix = user_movie_matrix.fillna(0)

# Transpose: now rows = movies, columns = users
movie_user_matrix_T = movie_user_matrix.T
```

```python
# Compute cosine similarity between movies
cosine_sim = cosine_similarity(movie_user_matrix_T)

# Put into a DataFrame for easy use
similarity_df = pd.DataFrame(cosine_sim, index=movie_user_matrix_T.index, columns=movie_user_matrix_T.index)
```

Function to Recommend Similar Movies

```python
def recommend_movies(movie_name, top_n=5):
    if movie_name not in similarity_df:
        return f"'{movie_name}' not found in dataset."
    
    sim_scores = similarity_df[movie_name].sort_values(ascending=False)
    return sim_scores[1:top_n+1]
```

```python
print(recommend_movies("Toy Story (1995)", top_n=5))
```
